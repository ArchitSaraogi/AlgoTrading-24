```python
import datetime as dt
import cryptocompare

start_date = dt.datetime(2020, 1, 1)
end_date = dt.datetime(2023, 12, 31)

# Fetch historical price data
data = []
current_date = start_date
while current_date < end_date:
    try:
        # Fetch daily data for the next 2000 days or until the end_date
        hist_data = cryptocompare.get_historical_price_day('ETH', 'USDT', limit=2000, toTs=int(current_date.timestamp()))
        if not hist_data:
            print("No more data available.")
            break
        data.extend(hist_data)
        
        # Update current_date to the date of the last fetched data point + 1 day
        last_date = dt.datetime.fromtimestamp(hist_data[-1]['time'])
        current_date = last_date + dt.timedelta(days=1)
        
        print(f"Fetching Data for {current_date}")
    except Exception as e:
        print(f"Error fetching data: {e}")
        break
```

```python
import pandas as pd
```

```python
# Create a DataFrame
df = pd.DataFrame(data)
df = df.drop_duplicates()

# Convert the timestamp to datetime
df['time'] = pd.to_datetime(df['time'], unit='s')

# Sort the DataFrame by time
df = df.sort_values('time')

# Set the datetime as the index
df.set_index('time', inplace=True)
df=df[2001:]

# Save the DataFrame to a CSV file
```

```python
df.to_csv('eth_usdt_2021_2023.csv')
```
